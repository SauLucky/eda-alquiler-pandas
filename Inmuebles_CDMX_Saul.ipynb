##DataFrame

Una estructura de datos fundamental en la biblioteca Pandas. Puedes imaginarlo como una tabla, similar a una hoja de cálculo de Excel o una tabla en una base de datos SQL. Aquí te detallo sus características principales:
* Estructura Tabular: Organiza los datos en filas y columnas, donde cada columna puede tener un tipo de dato diferente (números, texto, fechas, etc.).
* Etiquetas en Filas y Columnas: Cada fila y columna tiene una etiqueta (índice) que permite acceder y manipular los datos de manera eficiente. Por defecto, las filas tienen etiquetas numéricas (0, 1, 2, ...), pero se pueden personalizar.
* Tipos de Datos Heterogéneos: A diferencia de las matrices en NumPy, un DataFrame puede contener columnas con diferentes tipos de datos. Por ejemplo, una columna puede contener números enteros, otra cadenas de texto y otra fechas.
* Flexibilidad y Funcionalidades: Pandas proporciona una gran cantidad de funciones para limpiar, transformar, analizar y manipular los datos en un DataFrame. Puedes realizar operaciones como filtrar filas, seleccionar columnas, agrupar datos, realizar cálculos estadísticos, etc.

En resumen, un DataFrame es una herramienta poderosa y flexible para trabajar con datos tabulares en Python. Facilita la manipulación y el análisis de datos, y es esencial para muchas tareas de ciencia de datos y análisis exploratorio.

##figsize

figsize es un parámetro que se utiliza en las funciones de graficación de pandas (que a su vez se basa en la librería matplotlib) para especificar el tamaño de la figura donde se dibujará el gráfico.

El valor que le asignaste, (12, 8), es una tupla donde:

El primer número, 12, representa el ancho de la figura.
El segundo número, 8, representa la altura de la figura.
Estas unidades suelen estar en pulgadas cuando se trabaja con matplotlib.

Al especificar figsize=(12, 8), le estás diciendo a Pandas (y a Matplotlib por debajo) que quieres que el área donde se dibuje tu gráfico de barras horizontales tenga un ancho de 12 pulgadas y una altura de 8 pulgadas. Esto ayuda a controlar las dimensiones visuales de la gráfica, permitiendo que se vea más grande o más pequeña según tus necesidades, y asegurando que los elementos (como las etiquetas de las barras) se muestren correctamente sin superponerse.

En resumen, figsize te da control sobre las dimensiones de la imagen donde se dibuja tu gráfico.

### Ejemplo 1: Gráfico más pequeño

````python
# Gráfico con un tamaño más pequeño
df_tipo_precio.plot(kind='barh', figsize=(6, 4), color='purple')
````

### Ejemplo 2: Gráfico más ancho

````python
# Gráfico más ancho
df_tipo_precio.plot(kind='barh', figsize=(15, 8), color='purple')
````

### Ejemplo 3: Gráfico más alto

````python
# Gráfico más alto
df_tipo_precio.plot(kind='barh', figsize=(12, 10), color='purple')
````

Como puedes ver, ajustando los valores en la tupla `figsize=(ancho, alto)`, puedes modificar las dimensiones de la figura donde se dibuja el gráfico para que se adapte mejor a tus necesidades de visualización.

El equipo de desarrollo del sitio web de la empresa solicitó la creación de dos nuevas columnas numéricas en nuestra base de datos. En el sitio web de las propiedades, desean presentar información más resumida en relación a los valores. Para esto, estas columnas serán importantes:



### 1.-Valor_mensual: Esta columna debe contener los gastos mensuales de cada propiedad, incluyendo el alquiler y el condominio.


````python
datos = pd.read_csv(url, sep=';') #Almacenamos en la variable 'datos'
datos.shape
````

````python
datos['Valor_mensual'] = datos['Valor'] + datos['Condominio'] #Cremos la columna 'Valor_Mensual' que es 'Valor'+'Condominio'
datos.head()
````


### 2.- valor_anual: Esta columna debe contener los gastos anuales por propiedad, es decir, el Impuesto más 12 meses de alquiler y condominio.

````python
datos['Valor_anual'] = datos['Valor_mensual'] * 12 + datos['Impuesto'] #El Valor_anual va a tener el Impuesto mas los 12 meses del valor mensual.
datos.tail()
````

## **Crear columnas categóricas**

El equipo de desarrollo solicitó la creación de dos columnas adicionales. Sin embargo, esta vez son categóricas:



### 1.- Descripcion: Esta columna debe contener un resumen de la información clave de las propiedades que se mostrarán en el sitio web: tipo de propiedad, barrio, cantidad de habitaciones y plazas de estacionamiento.



````python
#Creamos la columna descripcion que contendra los valores concatenados de varias columnas tipo STRING, cadenas de texts, espacios,
#saltos de line y algunos INT convertidos a STR con el metodo .astype
datos['Descripcion'] = datos['Tipo'] + ' en la colonia ' + datos['Colonia'] + \
                     ' con ' + datos['Habitaciones'].astype(str) + ' cuarto(s) y ' + \
                     datos['Garages'].astype(str) + ' plazas de estacionamiento'
````

````python
datos.iloc[500].Descripcion #Localizamos el indice 500 y mostramos la Descricion
````


### 2.- Tiene_suite: Esta debe ser una columna que indique únicamente si la propiedad tiene o no suites, sin importar la cantidad.

````python
#Vamos a tomar la columna de suites y le vamos a aplicar una funcion anonima que va a verificar
#si exsite una suite y va dar como resultado un SI o un NO
datos['Tiene_suite'] = datos['Suites'].apply(lambda x: 'Si' if x > 0 else 'No')
````
